```python
import google.generativeai as genai
import json

# Configure Gemini
GOOGLE_API_KEY = "AIzaSyARFcY-llXeriWQI3Fs9Y8Kg6qBg0Xcipc"
genai.configure(api_key=GOOGLE_API_KEY)

# Create the model
model = genai.GenerativeModel(model_name="gemini-2.0-flash")

def convert_height(value, unit):
    if unit in ['cm', 'centimeters']:
        return value / 100
    elif unit in ['m', 'meters']:
        return value
    elif unit in ['ft', 'feet']:
        return value * 0.3048
    elif unit == 'in':
        return value * 0.0254
    return None

def convert_weight(value, unit):
    if unit in ['kg', 'kilograms']:
        return value
    elif unit in ['lb', 'lbs', 'pounds']:
        return value * 0.453592
    elif unit in ['stone', 'stones']:
        return value * 6.35029
    return None

def extract_measurements(json_data):
    height_m = 0
    weight_kg = 0

    for height_part in json_data['height']:
        value = height_part['value']
        unit = height_part['unit']
        height_m += convert_height(value, unit)

    weight_data = json_data['weight']
    weight_kg += convert_weight(weight_data['value'], weight_data['unit'])

    return round(height_m, 2), round(weight_kg, 2)

def process_input(text):
    prompt = f"""
Extract the height and weight from this sentence.
- If height is given in feet and inches, separate them into two entries.
- Output ONLY strict JSON with this format:
{{
    "height": [
        {{"value": number, "unit": "unit"}},
        {{"value": number, "unit": "unit"}}
    ],
    "weight": {{"value": number, "unit": "unit"}}
}}

Text: "{text}"
"""

    response = model.generate_content(prompt)
    result = response.text.strip()

    # Try to safely load JSON
    try:
        json_result = json.loads(result)
    except json.JSONDecodeError as e:
        raise ValueError(f"Failed to parse JSON from Gemini output: {result}") from e

    return json_result

# Example usage
user_input = "I'm about 5'10 and weigh 150 pounds"

try:
    data = process_input(user_input)
    print("Extracted JSON:", data)
    height, weight = extract_measurements(data)
    print(f"Height: {height} meters, Weight: {weight} kg")
except Exception as e:
    print("Error:", str(e))
```
